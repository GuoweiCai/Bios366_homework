```python
"""
Different cooling schedule
"""
periods = 15
tau_start = 10
neighbor = 1

# Cooling schedule_1
tau_1 = [tau_start * 0.85**i for i in range(15)]

# Cooling schedule_2
tau_2 = [tau_start * 0.90**i for i in range(15)]

# Cooling schedule_3
tau_3 = [tau_start * 0.95**i for i in range(15)]

annealing(periods, tau_1, neighbor, 'Cooling schedule 1 (0.85**i)')
annealing(periods, tau_2, neighbor, 'Cooling schedule 2 (0.90**i)')
annealing(periods, tau_3, neighbor, 'Cooling schedule 3 (0.95**i)')
```

```python
"""
2.1
GA--Different mutation rates
"""

def GA(pop_size, iterations, mutation_rate, Title, independence = None):
    
    aic = lambda g: g.nobs * np.log((g.resid**2).sum()/g.nobs) + 2*len(g.beta)
    aic_best = []
    best_solution = []
    aic_history = []
    
    # Initialize genotype
    current_gen = np.random.binomial(1, 0.5, pop_size*ncols).reshape((pop_size, ncols))
    
    def calculate_fitness(aic_values):
        P = len(aic_values)
        aic_rank = (-aic_values).argsort().argsort()+1.
        return 2.*aic_rank/(P*(P+1.))
    
    for i in range(iterations):
        
        # Get phenotype
        current_phe = [predictors[predictors.columns[g.astype(bool)]] for g in current_gen]
        # Calculate AIC
        current_aic = np.array([aic(pd.ols(y=logsalary, x=x)) for x in current_phe])
        # Get lowest AIC
        aic_best.append(current_aic[np.argmin(current_aic)])
        best_solution.append(current_gen[np.argmin(current_aic)])
        
        # Calculate fitness according to AIC rank
        fitness = calculate_fitness(current_aic)
        
        # Choose first parents according to fitness
        moms = np.random.choice(range(pop_size), size=pop_size/2, p=fitness)
        if independence:
            # Choose first parents according to fitness
            dads = np.random.choice(range(pop_size), size=pop_size/2, p=fitness)
        else:
            # Choose second parents randomly
            dads = np.random.choice(range(pop_size), size=pop_size/2, p=None)
        
        next_gen = []
        for x,y in zip(current_gen[moms], current_gen[dads]):
            # Crossover
            cross = np.random.randint(0, ncols)
            child1 = np.r_[x[:cross], y[cross:]]
            child2 = np.r_[y[:cross], x[cross:]]
            # Mutate
            m1 = np.random.binomial(1, mutation_rate, size=ncols).astype(bool)
            child1[m1] = abs(child1[m1]-1)
            m2 = np.random.binomial(1, mutation_rate, size=ncols)
            child2[m2] = abs(child1[m2]-1)
            next_gen += [child1, child2]
            
        # Increment generation
        current_gen = np.array(next_gen)
        # Store AIC values
        aic_history.append(current_aic)
    plt.figure()
    plt.title(Title)
    plt.plot(aic_best)
    plt.figure()
    plt.title(Title)
    for i,x in enumerate(aic_history):
        plot(np.ones(len(x))*i, -x, 'r.', alpha=0.3)
```

```python
"""
Different muation rates
"""

pop_size = 20
iterations = 100


# Mutation rates 1
mutation_rate_1 = .20

# Mutation rates 2
mutation_rate_2 = .10

# Mutation rates 3
mutation_rate_3 = .05


GA(pop_size, iterations, mutation_rate_1, 'Mutation_rate : .20')
GA(pop_size, iterations, mutation_rate_2, 'Mutation_rate : .10')
GA(pop_size, iterations, mutation_rate_3, 'Mutation_rate : .05')
```

```python
"""
2.2
Different generation size
"""

mutation_rate = .20
iterations = 100


# Mutation rates 1
pop_size_1 = 20

# Mutation rates 2
pop_size_2 = 50

# Mutation rates 3
pop_size_3 = 80


GA(pop_size_1, iterations, mutation_rate, 'Generation size : 20')
GA(pop_size_2, iterations, mutation_rate, 'Generation size : 50')
GA(pop_size_3, iterations, mutation_rate, 'Generation size : 80')
```
